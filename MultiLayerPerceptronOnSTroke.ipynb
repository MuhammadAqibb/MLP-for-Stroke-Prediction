```python
import pandas as pd
import numpy as np
#from keras.models import load_model
from sklearn.neural_network import MLPClassifier
#from keras.models import Sequential
#from keras.layers import Dense
training_data = pd.read_csv('/content/drive/MyDrive/MLP_on_stroke/StrokeDataTrain.csv')
```

Shuffle Data

```python
#training_data = training_data.sample(frac=1).reset_index(drop=True)
#training_data.to_csv('/content/drive/MyDrive/MLP_on_stroke/StrokeDataTrain.csv', index=False)
#training_data.head(5)
```

Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
columns_to_encode = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']
for column in columns_to_encode:
  encoded_column = encoder.fit_transform(training_data[column])
  training_data[column] = encoded_column
```

```python
X = training_data.iloc[:,:-1].values
Y = training_data.iloc[:,-1].values
```

removing null in the data

```python
training_data['bmi'].fillna(training_data['bmi'].mean(), inplace=True)
#check whether null has been replaced with mean
training_data.isnull().sum()
```

Scaling data to optimum values

```python
#from sklearn.preprocessing import StandardScaler
#sc = StandardScaler()
#X = sc.fit_transform(X)
```

```python
X.head(5)
```

Test_train_split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

Multilayer perceptron model

```python
#model = Sequential()
#model = MLPClassifier()
model = MLPClassifier(hidden_layer_sizes=(10,), max_iter=1000)
model.fit(X_train, Y_train)
#model.add(Dense(units=64, activation='relu', input_dim=X_train.shape[1])) #64 neurons, input dimensions = X train
#model.add(Dense(units=32, activation='relu'))
#model.add(Dense(units=1, activation='sigmoid'))
```

compilation, Optimization with adam optimizer, loss function = binary crossentropy (good for classification) 

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

train the model on data

```python
model.fit(X_train, Y_train, epochs=100, batch_size=32, validation_data=(X_test, Y_test))
```

```python
score = model.evaluate(X_test, Y_test)
print("Accuracy: %.2f%%" % (score[1]))
```
